# MP neuron model

```
# Let us see for different values of i
# generates random integer

from random import randint 
```

```
# The model of the MP neuron is single parameter (single parametric model)
# Given a particular data item (binarized training data set) we add up all the 
# features which are in the binary format and see if the sum exceeds the value 
# of the single parameter 'b'
# If it exceeds b or equals b then we say the MP neuron model outputs 1 (malignant)
# If it does not exceeds then the MP neuron model outputs 0 (begnin)

# Let us see with a particular parameter say b as 3

# But our goal is to find out the value of b such that the accuracy of finding
# the actual value across all the rows by the model is maximized (or) the loss is minimum

b = 3 # Let us say the single parameter is initialized as 3

#i = 100 # 100 th row of the training data
i = randint(0, X_binarised_train.shape[0])
print('For row', i)

# Predicted value by the model
if(np.sum(X_binarised_train[i]) >= b):
  print("MP neuron inference is malignant")
else:
  print("MP neuron inference is benign")

# Actual value present in the data set
if(Y_train[i] == 1):
  print("Ground truth is malignant")
else:
  print("Ground truth is benign")
```

```
# Let us first take a specific value of b and compute the accuracy over all rows
b = 3 # initialse b as 3

Y_pred_train = [] # empty list or 1D array that will store the predicted values
accurate_rows = 0 # keeps the count correctly predicted rows

# zip function is used to do two iterations parallely at the same time
for x, y in zip(X_binarised_train, Y_train):
  # iterating x over X_binarised_train
  # iterating y over Y_train
  # x is 30 features or one row in X_binarised_train
  # y is the associated label or target of that row
  # sum of 30 features of a single row compared to b 
  # which returns a boolean value stored in a variable
  y_pred = (np.sum(x) >= b)
  # boolean value is appended to the list Y_pred_train
  Y_pred_train.append(y_pred)
  accurate_rows += y == y_pred
# X_binarised_train.shape[0] returns the number of rows value as
# X_binarised_train.shape is tuple data type
#print(type(y_pred))
#print(type(y))
print("accurate # prediction: ", accurate_rows, ",", "accuracy: ", accurate_rows/X_binarised_train.shape[0])
# The accuracy is very low because we only get 15% accuracy since the data has 
# 569 rows out of which only 15% of 569 rows were correctly predicted. This is 
# because we use the only one bias which is b = 3 that outputs "malignant" always.
# Actually there are 63 % of data being malignant (refer data.groupby('class').mean() for mean of malignant) out of which only 15% is predicted
# correctly when b is 3 
''' find this "data['class'].value_counts()" and this "data.describe()" in this notebook to see the mean value of class malignant'''
```

```
# Let us the redo the program for all values of 'b'.
# We want to obtain such a value of 'b' such that the accuracy over all the 
# training dataset is maximized

# X_binarised_train.shape[1] returns the # of columns value since it is a 
# tuple data type and the reason for X_binarised_train.shape[1] + 1 is that 
# the bias value will also include from 0 to X_binarised_train.shape[1]
# or o to # columns of training data + 1
# i.e., [0, X_binarised_train.shape[1]]
for b in range(X_binarised_train.shape[1] + 1):
  Y_pred_train = [] # empty list or 1D array that will store the predicted values
  accurate_rows = 0 # keeps the count correctly predicted rows

  # zip function is used to do two iterations parallely at the same time
  for x, y in zip(X_binarised_train, Y_train):
    # iterating x over X_binarised_train
    # iterating y over Y_train
    # x is 30 features or one row in X_binarised_train
    # y is the associated label or target of that row
    y_pred = (np.sum(x) >= b) # sum of 30 features of a single row compared to b and stored in a variable
    Y_pred_train.append(y_pred) # stored varialbe is appended to the list Y_pred_train
    accurate_rows += y == y_pred
    # X_binarised_train.shape[0] returns the number of rows value as
    # X_binarised_train.shape is tuple data type
    # X_binarised_train.shape[0] retuens the # rows
  print("Bias value: ", b, ", ", "Accuracy: ", accurate_rows/X_binarised_train.shape[0])
  # Accuracy is high when the value of b is 0 and then accuracy is reduced for all other values
  # of b for all the training data set
  # Actually there are 63 % of data being malignant out of which our MP neuron model manages
  # to predict it more or less same but only for bias  value 0
  # This means the accuracy reduces again for different values of b and ends up in a situation
  # where accuracy drastically reduces and ends up being poor
  # Remember our goal is to obtain such a value of 'b' such that the accuracy over all the 
  # training dataset is maximized and so the model classifies them correctly as possible as actual
```

```
# Why does this happen?
# Let us look into our data itself to see what needs to be improved to optimize
# our model for accurate prediction

''' Find this "data.groupby('class').mean()" in this notebook to see more about the data'''
# What we can vaguely observe in the data is that the features of the data of class 0 (benign) has very high value
# when compared to the features with class 1 (malignant)

# This means the problem that we face right now to improve the accuracy over all the data points 
# of the training data seems to be in the way we have binarised the inputs to the model
# Remember that class 0 is benign and class 1 is malignant and also we know that the mean of our 
# binary class label has around 63% of malignant data (All these leads us to make an inference that
# the model could improve its accuracy when we try to toggle the labels while binarising the inputs)

''' Refer this X_binarised_train = X_train.apply(pd.cut, bins = 2, labels = [0, 1]) and this
X_binarised_test = X_test.apply(pd.cut, bins = 2, labels = [0, 1]) to see how the binarisation of inputs or features happen'''
# That is whenever we are trying to split the dataset the above code seen within ''' ''' tries to find a threshold value in the features
# and assigns o to those values which are below that value and assigns 1 to the rest.
# And that tendency to label it will automatically goes to malignant as 1 because we did not observe the data (

''' Toggling the labels while binarising the input features'''
# But now that we have observed our dataset has very low value for malignant and comparitively high value for benign) we can try to change that
# and see whether our model maximizes the accuracy over all the training data set and try to find out such a value of 'b' that maximizes the 
# accuracy of the model over all different values of bias 'b' in [0, X_binarised_train.shape[1] + 1]

# Flipping the labels of feature binarisation for both training and testing data set
X_binarised_train = X_train.apply(pd.cut, bins = 2, labels = [1, 0]) # Flipped the labels because we observed different trend from actual dataset
X_binarised_test = X_test.apply(pd.cut, bins = 2, labels = [1, 0]) # Flipped the labels because we observed different trend from actual dataset
print(type(X_binarised_train))
print(type(X_binarised_test))
# We want to convert them into array in order for numerical computation 
# and so we have to convert from pandas data frame to numpy arrays
X_binarised_train = X_binarised_train.values
X_binarised_test = X_binarised_test.values
print(type(X_binarised_train))
print(type(X_binarised_test))
```

```
''' Let's re-run the code and see the accuracy of the model over all the training data set for different values of b in an attempt to
find out that the value of b which improves the accuracy of classification of the model '''

''' Training data (Otimization can be seen as search algorithm and inference (accuracy metrics) problem to find the optimal value) '''
# Let us the redo the program for all values of 'b'.
# We want to obtain such a value of 'b' such that the accuracy over all the 
# training dataset is maximized

# X_binarised_train.shape[1] returns the # of columns value since it is a 
# tuple data type and the reason for X_binarised_train.shape[1] + 1 is that 
# the bias value will also include from 0 to X_binarised_train.shape[1]
# or o to # columns of training data + 1
# i.e., [0, X_binarised_train.shape[1]]
for b in range(X_binarised_train.shape[1] + 1):
  Y_pred_train = [] # empty list or 1D array that will store the predicted values
  accurate_rows = 0 # keeps the count correctly predicted rows

  # zip function is used to do two iterations parallely at the same time
  for x, y in zip(X_binarised_train, Y_train):
    # iterating x over X_binarised_train
    # iterating y over Y_train
    # x is 30 features or one row in X_binarised_train
    # y is the associated label or target of that row
    y_pred_ = (np.sum(x) >= b) # sum of 30 features of a single row compared to b and stored in a variable
    Y_pred_train.append(y_pred_) # stored varialbe is appended to the list Y_pred_train
    accurate_rows += y == y_pred_
    # X_binarised_train.shape[0] returns the number of rows value as
    # X_binarised_train.shape is tuple data type
    # X_binarised_train.shape[0] retuens the # rows
  print("Bias value: ", b, ", ", "Accuracy: ", accurate_rows/X_binarised_train.shape[0])

  # So what we see in the result is that the accuracy has consistently increased over the training data set
  # and the value of bias or b at 28 gives an accuracy of around 85% which is atleast way better than the previous results
  # that was obtained without observing and binarising the trend of the data

''' Therefore binarising of the input features plays a key role in increasing the accuracy of the training data '''
```

```
''' MP  neuron model - Test data (Inference problem (accuracy metrics)) '''

# Let us try to test our MP neuron model accuracy when the bias 'b' is at 28
# We can compute accuracy using scikit learn instead of manual computation

from sklearn.metrics import accuracy_score
b = 28 # the bias that gave highest accuracy for the model in the training data
Y_pred_test = [] # empty list or 1D array
for x in X_binarised_test: # iterate over all each test data
  y_pred = (np.sum(x) >= b) # y_pred stores the scalar value summation of 30 features computed and compared with b value
  Y_pred_test.append(y_pred) # stores the y_pred value computed as a list for each test data
  ''' The function of accuracy_score is to return a float value corresponding to the
  similarities between true labels and predicted labels'''
accuracy = accuracy_score(Y_pred_test, Y_test) # computes the accuracy
print("Bias: ", b, ",", "Accuracy: ", accuracy)

# What we can observe here is that the accuracy of the MP neuron model on the test data is low when compared to
# the accuracy acquired in the training data
# This is not in our control (we can only train our model through training data and try to search and infer the best value (optimal value) of
# b or bias for training data alone. But the behavior that this bias or b exhibits for the test data is not in our hand for improvisation
# i.e., we have control only over training data so that we can try to improvise our models' accuracy whereas in the testing part we are
# only inferring how the trained model behaves or how much accuracy does it exhibits for unseen data)
```

# MP Neuron and Perceptron using Python: Class

```
# Using Class to templatize the model
class MPNeuron:
  
  # Initialization function
  def __init__(self):
    # we define the parameters of the model
    self.b = None
  
  # Inference function  
  def model(self, x):
    # returns true if 1 or false if 0
    return(sum(x) >= self.b)

  # Predict function
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)    
    #print(Y)
    return np.array(Y)

  # Fit function - to find the right value of b from all possible options
  # learning algorithm
  def fit(self, X, Y):
    accuracy = {}
    for b in range(X.shape[1] + 1):
      self.b = b
      Y_pred = self.predict(X)
      accuracy[b] = accuracy_score(Y_pred, Y)
    best_b = max(accuracy, key = accuracy.get)
    self.b = best_b

    print("Optimal value of b: ", best_b)
    print("Highest accuracy: ", accuracy[best_b])
```

```
# How to use the MP Neuron class model
mp_neuron = MPNeuron()
mp_neuron.fit(X_binarised_train, Y_train)
```
