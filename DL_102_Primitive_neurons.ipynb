# MP neuron model

```
# Let us see for different values of i
# generates random integer

from random import randint 
```

```
# The model of the MP neuron is single parameter (single parametric model)
# Given a particular data item (binarized training data set) we add up all the 
# features which are in the binary format and see if the sum exceeds the value 
# of the single parameter 'b'
# If it exceeds b or equals b then we say the MP neuron model outputs 1 (malignant)
# If it does not exceeds then the MP neuron model outputs 0 (begnin)

# Let us see with a particular parameter say b as 3

# But our goal is to find out the value of b such that the accuracy of finding
# the actual value across all the rows by the model is maximized (or) the loss is minimum

b = 3 # Let us say the single parameter is initialized as 3

#i = 100 # 100 th row of the training data
i = randint(0, X_binarised_train.shape[0])
print('For row', i)

# Predicted value by the model
if(np.sum(X_binarised_train[i]) >= b):
  print("MP neuron inference is malignant")
else:
  print("MP neuron inference is benign")

# Actual value present in the data set
if(Y_train[i] == 1):
  print("Ground truth is malignant")
else:
  print("Ground truth is benign")
```

```
# Let us first take a specific value of b and compute the accuracy over all rows
b = 3 # initialse b as 3

Y_pred_train = [] # empty list or 1D array that will store the predicted values
accurate_rows = 0 # keeps the count correctly predicted rows

# zip function is used to do two iterations parallely at the same time
for x, y in zip(X_binarised_train, Y_train):
  # iterating x over X_binarised_train
  # iterating y over Y_train
  # x is 30 features or one row in X_binarised_train
  # y is the associated label or target of that row
  # sum of 30 features of a single row compared to b 
  # which returns a boolean value stored in a variable
  y_pred = (np.sum(x) >= b)
  # boolean value is appended to the list Y_pred_train
  Y_pred_train.append(y_pred)
  accurate_rows += y == y_pred
# X_binarised_train.shape[0] returns the number of rows value as
# X_binarised_train.shape is tuple data type
#print(type(y_pred))
#print(type(y))
print("accurate # prediction: ", accurate_rows, ",", "accuracy: ", accurate_rows/X_binarised_train.shape[0])
# The accuracy is very low because we only get 15% accuracy since the data has 
# 569 rows out of which only 15% of 569 rows were correctly predicted. This is 
# because we use the only one bias which is b = 3 that outputs "malignant" always.
# Actually there are 63 % of data being malignant (refer data.groupby('class').mean() for mean of malignant) out of which only 15% is predicted
# correctly when b is 3 
''' find this "data['class'].value_counts()" and this "data.describe()" in this notebook to see the mean value of class malignant'''
```

# MP Neuron and Perceptron using Python: Class

```
# Using Class to templatize the model
class MPNeuron:
  
  # Initialization function
  def __init__(self):
    # we define the parameters of the model
    # single parameter 'b' for MP Neuron
    self.b = None
  
  # Inference function
  # Takes the input data and predicts the output  
  def model(self, x):
    # returns true if 1 or false if 0
    return(sum(x) >= self.b)

  # Predict function
  # Takes entire set of input data rows and predict the corresponding Y output
  def predict(self, X):
    Y = []
    for x in X:
      result = self.model(x)
      Y.append(result)    
    #print(Y)
    return np.array(Y)

  # Fit function - to find the right value of b from all possible options
  # learning algorithm
  def fit(self, X, Y):
    accuracy = {}
    for b in range(X.shape[1] + 1):
      self.b = b
      Y_pred = self.predict(X)
      accuracy[b] = accuracy_score(Y_pred, Y)
    best_b = max(accuracy, key = accuracy.get)
    self.b = best_b

    print("Optimal value of b: ", best_b)
    print("Highest accuracy: ", accuracy[best_b])
```

```
# How to use the MP Neuron class model
mp_neuron = MPNeuron()
mp_neuron.fit(X_binarised_train, Y_train)
```

# Perceptron model

```
# To smoothen the graph between epoch ans accuracy is through learning rate
class Perceptron:
  # initialize parameters
  def __init__(self):
    # w and b are weights and bias
    self.w = None # w is an array vector
    self.b = None # b is a scalar

  # Model for a single given input (given an x predict y)
  # Inference problem
  def model(self, x):
    return 1 if (np.dot(self.w, x) >= self.b) else 0


  # Model for a entire set of input (given X predict Y) or 
  # takes vector of x as input and predicts vector of Y as output
  # Inference problem
  def predict(self, X):
    Y = [] # vector 
    for x in X:
      result = self.model(x)
      Y.append(result) # list of scalars or list vector (slow)
    return np.array(Y) # array vector (fast) -> Y predicted

  # given X and Y learn the right parameters    
  def fit(self, X, Y, epochs = 1, lr = 1):
    # Model has to deal with all possible combinations of w and b
    # Unlike MP Neuron model where 'b' is just few integers (#features + 1), 
    # whereas in Perceptron model 'b' is any real valued number and so is the 
    # case with 'w'. Therefore we cannot go for brute force search because it will
    # be inefficient.
    # Therefore we have learning algorithm for Perceptron model (w = w + x and 
    # w = w- x)
    # Learning problem
    # Inital values of w and b
    self.w = np.ones(X.shape[1])
    self.b = 0
    # For plotting training accuracy
    accuracy = {} # dictionary
    max_accuracy = 0
    for i in range(epochs):
      for x, y in zip(X, Y): # X train and Y train or actual or ground truth
        y_pred = self.model(x)
        # Learning occurs here (in case of error between actual and predicted y)
        if y == 1 and y_pred == 0: # cosine angle is obtuse
          # To make cosine angle acute (cosine value within 1 to 0)
          ''' learning happens in these statments (changing w and b by learning
          from data) and we like to reduce learning rate in order to improve
          the accuracy as well as learning pattern'''
          # learning rate - lr is a fraction less than 1
          self.w = self.w + lr * x 
          self.b = self.b + lr * 1 # bias can take any real value
        elif y == 0 and y_pred == 1: # cosine angle is acute
          # To make cosine angle obtuse (cosine value within 0 to -1)
          ''' learning happens in these statments (changing w and b by learning
          from data) and we like to reduce learning rate in order to improve
          the accuracy as well as learning pattern'''
          # learning rate - lr is a fraction less than 1
          self.w = self.w - lr * x
          self.b = self.b - lr * 1 # bias can take any real value


      # computes accuracy after each epoch    
      accuracy[i] = accuracy_score(self.predict(X), Y)
      #print(accuracy[i])
      if (accuracy[i] > max_accuracy):
        max_accuracy = accuracy[i]
        # Whenever there is an increase in accuracy we store those parameter values
        chkptw = self.w
        chkptb = self.b 
    # When model ends at a particular point we ensure that it ends up where
    # model parameters has the highest training accuracy (Remember we do not
    # have control over test data i.e., we are only maximizing the training accuracy)    
    self.w = chkptw
    self.b = chkptb
    print(max_accuracy)    
    #plt.plot(accuracy.values())
    # To plot input to be passed for accuracy.values() and loss .values() is of 
    # type list() when using dictionary for accuracy and loss
    plt.plot(list(accuracy.values()))
    plt.ylim([0, 1]) # accuracy number setting as 0 to 1
    plt.show()
```
